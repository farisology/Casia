```python
import os
import openai
import pandas as pd

pdb = pd.read_csv("plantDB abc.csv")

pdb
```

```python
def open_file(filepath):
    with open(filepath, 'r', encoding='utf-8') as infile:
        return infile.read()

start_sequence = "CleanText:"

def gpt_janitor(prompt):
    response = openai.Completion.create(
      model="text-davinci-003",
      prompt= prompt,
      temperature=0,
      max_tokens=1000,
      top_p=1,
      frequency_penalty=0,
      presence_penalty=0
    )
    return response["choices"][0]["text"]
```

```python
openai.api_key = open_file('openai_key.txt')

clean_plant_details = list()

for index, row in pdb.iterrows():
    new_prompt = open_file('casia_prompt.txt').replace('<<RAW_TEXT>>', row[3])
    clean_plant_details.append(gpt_janitor(new_prompt))
    # break
```

```python
pdb["clean_deatils"] = clean_plant_details
```

```python
pdb.to_csv("clean plant DB.csv", index=False)
pdb
```
