```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import psycopg2
```

```python
# URL to scrap
def scrape_post(post_path):
    # URL to scrape
    url = post_path

    # Send a GET request to the URL and get the response object
    response = requests.get(url)

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, "html.parser")

    # Find all div tags with class name "notebody"
    # note_subheaders = soup.find_all("h3")
    notebodies = soup.find_all("p")
    
    content_list = list()
    # Loop through each notebody div tag and extract its inner text
    for notebody in notebodies:
        content_list.append(notebody.text)
    return '\n'.join(content_list)

data = []

for num in range(1,16):
    url = f"https://www.houseplant411.com/houseplant/page/{num}"
    
    # Send a GET request to the URL and get the response object
    response = requests.get(url)
    
    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, "html.parser")
    
    for post in soup.find_all("h2", {"class": "entry-title"}):
        title = post.find("a").text.strip()
        post_link = post.find("a")["href"]
        # print(title)
        # print(post_link)

        try:
            content = scrape_post(post_link)
            print(title, "Successful!!")
        except Exception as e:
            continue
        row = {
            "base_url": url,
            "title": title,
            "post_link": post_link,
            "post_date": "",
            "tags": "",
            "post_content": content
        }
        
        data.append(row)
    # break
```

```python
plant_411_df = pd.DataFrame(data)

plant_411_df
```
